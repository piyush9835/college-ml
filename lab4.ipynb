```python
import pandas as pd
import numpy as np


# Replace 'yourfile.csv' with the name of your uploaded file
df = pd.read_csv('titanic.csv')

# Display the first few rows
df.head()
```

```python
missing_values = df.isnull().sum()
print("Missing values:\n", missing_values)
```

```python
survived_count = df['survived'].sum()
print("Number of people who survived:", survived_count)
```

```python
survived_gender = df[df['survived'] == 1]['sex'].value_counts()
print("Number of males and females who survived:\n", survived_gender)
```

```python
average_age_survived = df[df['survived'] == 1]['age'].mean()
print("Average age of people who survived:", average_age_survived)
```

```python
min_age = df[df['survived'] == 1]['age'].min()
max_age = df[df['survived'] == 1]['age'].max()

print("Minimum age of survivors:", min_age)
print("Maximum age of survivors:", max_age)
```

```python
survival_by_class = df.groupby('class')['survived'].sum()
print("Survival count by class:\n", survival_by_class)
```

```python
passengers_with_sibsp = df[df["SibSp"] > 0].shape[0]
print("Number of passengers with siblings/spouses:", passengers_with_sibsp)
```

```python
passengers_with_parch = df[df["Parch"] > 0].shape[0]
print("Number of passengers with parents/children:", passengers_with_parch)
```

```python
last_col_name = df.columns[-1]
print("Last column name:", last_col_name)
```

```python
first_col_name = df.columns[1]
print("First column name:", first_col_name)
```
